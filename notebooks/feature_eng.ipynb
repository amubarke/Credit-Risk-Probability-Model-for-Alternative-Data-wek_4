```python
from data_processing import build_pipeline
import pandas as pd
categorical_cols = ["ProductCategory", "ChannelId"]
numeric_cols = ["Amount"]

data = pd.read_csv(r"E:\Credit-Risk-Probability-Model-for-Alternative-Data-wek_4\data\raw\data.csv")
categorical_cols = data.select_dtypes(include=["object", "category"]).columns.tolist()

# Build the pipeline
pipeline = build_pipeline(
    categorical_cols=categorical_cols,
    numeric_cols=numeric_cols,
    target_col="FraudResult",
    scaling_method="standard"
)
# Fit & transform the data
processed_data = pipeline.fit_transform(data, data["FraudResult"])

print("Processed data shape:", processed_data.shape)
```
